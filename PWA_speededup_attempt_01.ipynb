```julia
noϵ = [i==1 for i=1:size(waves[:,6],1)]
posϵ = [ϵ=="+" for ϵ in waves[:,6]]
negϵ = [ϵ=="-" for ϵ in waves[:,6]]
## decomp
sum(noϵ)+sum(posϵ)+sum(negϵ)==length(waves[:,6])
block(i) = 1*noϵ[i]+2*posϵ[i]+3*negϵ[i]
```

```julia
COH = [block(i)==block(j) for i=1:88, j=1:88];
const COHc = convert(Array{Complex{Float64},2},COH);
```

```julia
function COHsq(X::AbstractVector{Complex{Float64}})
    sum(abs2(sum(X[bl])) for bl in [noϵ,posϵ,negϵ])
end
```

```julia
const ModelBlocks = [noϵ,posϵ,negϵ,negϵ]
Trel = let vect=[]
    temp = []
    numb = []
    for bl in ModelBlocks
        push!(temp,false)
        push!(temp,[true for i=1:(sum(bl)-1)]...)
        push!(numb,collect(1:88)[bl]...)
    end 
    Trel = zeros(Complex{Float64},88,sum(temp+1))
    global Tmap = fill(0,2,sum(temp+1))
    count=1
    for (i,b) in enumerate(temp)
        Trel[numb[i],count] = 1.0;
        Tmap[1,count] = numb[i]
        count+=1
        if b
            Trel[numb[i],count] = 1.0im;
            Tmap[2,count] = numb[i];
            count+=1
        end
    end
    Trel
#     numb
end;
```

```julia
function TTm(p)
    c = fill(0.0+0.0im,88)
    for i in 1:size(Tmap,2)
        Tmap[1,i]!=0 && (c[Tmap[1,i]]+=1.0*p[i])
        Tmap[2,i]!=0 && (c[Tmap[2,i]]+=1.0im*p[i])
    end
    c
end
```

Likelihood functions

```julia
function LLH(pars::Vector{Float64})
    res = sum(log(real(begin
#                     m = get_form(BF[e,:])
                    cv = BF[e,:].*(TT*pars)
                    COHsq(cv)
                    end)) for e in 1:Nd)
    - res + real(pars'*TBmatT*pars) * Nd
end
function dLLH(pars::Vector{Float64})
    resP = sum(2real(begin
                    m = get_form(BF[e,:])
                    v = m*pars
                    v / (pars'*v)
                end)  for e in 1:Nd)
    - resP + 2real(TBmatT*pars)* Nd 
end
function LLH_and_GRAD!(pars::Vector{Float64}, grad::Vector{Float64})
    val = 0.0; grad .= 0.0;
    for e in 1:Nd
        m = get_form(BF[e,:])
        v = m*pars
        sc = real(pars'*v)
        val -= log(sc)
        grad .-= (2real(v) ./ sc)
    end
#     println(val)
    BB = TBmatT*pars;
    val += real(pars'*BB) * Nd;
    grad .+= 2real(BB)* Nd;
    return val;
end
```
